```python
def make_hist(d_info):
    "Выодит гистограму по полученным данным"
    plt.figure(figsize= (8,6))
    sns.histplot(d_info)
    plt.xticks(rotation = 90)
    plt.ylabel('Число опрошенных')
    plt.show()
    return
```

```python
def calc_percent(cnt, r = 2):
    """ Считает процентное соотношение числа опрошенных,всего опрошенных 100%
        r - окгугление до r знаков после запятой
    """
    s = sum(cnt.values())
    p_cnt = dict()

    for key in cnt.keys():
        p_cnt[key] = round(cnt[key] / s * 100, r)

    p_cnt['Всего'] = s
    return p_cnt
```

### Оцените качество официальной информации о состоянии конкурентной среды

```python
# Оцените качество официальной информации о состоянии конкурентной среды на рынках товаров и услуг ивановской области
from IPython.display import HTML, display
st = 15
end = 17

html_str = "<table> "
for i in range(st, end+1):
    html_str += f"<tr> <td> {kin_data.columns[i]} </td>"
    ans = dict(Counter(kin_data.iloc[:, i]).most_common())
    for key in ans.keys():
        html_str += f"<td> {key} </td>  <td> {ans[key]} </td>"
    html_str += f"</tr>"

html_str +='</table>'
display(HTML(html_str))
#print(html_str)
```

```python
def fill_freq_mat(data, cats, columns):
    N = data.shape[0]
    M = data.shape[1]

    freq_mat = {}
    for col in cats:  # создаем матрицу для подсчета категорий
        freq_mat[col] = np.zeros(len(columns))

    for i in range(N):
        for j in range(M):
            freq_mat[data.iloc[i,j]][j] += 1

    return pd.DataFrame(data = freq_mat, index = columns)
```

```python
# Тепловая карта
cat = ["Удовлетворительно", "Скорее удовлетворительно", "Скорее неудовлетворительно", "Затрудняюсь ответить/мне ничего не известно о такой информации"]

frame = kin_data.iloc[:, st:end+1]
frame.shape

# сокращенные названия строк вопроса анкеты
tr = r'Оцените качество официальной информации о состоянии конкурентной среды на рынках товаров и услуг Ивановской области, размещаемой в открытом доступе: /'
columns = []
for col in frame.columns:
    columns.append(col[len(tr)+1:])

freq = fill_freq_mat(frame, cat, columns)


plt.figure(figsize = (18,9))
sns.heatmap(freq, vmin = 0, cmap ='viridis', linewidths= 1, annot = True, fmt ='0.1f')
```
